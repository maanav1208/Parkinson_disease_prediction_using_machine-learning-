# Analysis on the fundamental frequencies.. 

```python
#  For Average fundamental frequencies..


# box plot of affected_freq_avg and not_affected_freq_avg ..

affected_freq_avg = parkinson_data[parkinson_data['status']==1]['MDVP:Fo(Hz)'].values
not_affected_freq_avg = parkinson_data[parkinson_data['status']==0]['MDVP:Fo(Hz)'].values
plt.boxplot([affected_freq_avg,not_affected_freq_avg])
plt.title('Box Plots of Average vocal fundamental frequency per Parskinson Positive and Parskinson Negative cases')
plt.xticks([1,2],('Parkinson Positive cases','Parkinson Negative cases'))
plt.grid()
plt.show()
```
